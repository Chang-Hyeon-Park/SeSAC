## 피보나치 수열의 개념을 정리하고 코드를 수정작성하여 공유해봅시다.

피보나치 수열
 - 첫째와 둘째 항이 1이며, 그 뒤의 모든 항은 바로 앞 두 항의 합인 수열
 - 1 / 1 / 2(1+1) / 3(1+2) / 5(2+3) / 8(3+5)
 - 편의상 0번째 항을 0으로 두기도 한다.

````python
def fibo1(n):
  a, b = 0, 1
  if n==1 or n==2:
    return 1

  for i in range(n):
    a, b = b, a+b
  return a

num = int(input("n 번째의 피보나치 수열값을 보여드릴께요: "))
result = fibo1(num)
print(result)
````

## 가변인수(arbitrary argument)

가변 인수 : 함수가 임의의 수의 인수를 받을 수 있게 해줍니다.

1. 튜플 매개 인수(*args)

*args는 함수에 여러 개의 위치 인수를 전달할 때 사용

함수 내에서 args는 튜플로 취급되며, 전달된 모든 위치 인수를 포함

2. 딕셔너리 매개 인수(**kwargs)

**kwargs는 함수에 여러 개의 키워드 인수를 전달할 때 사용

함수 내에서 kwargs는 딕셔너리로 취급되며, 전달된 모든 키워드 인수를 포함.

````python
def print_args(*args):
  for arg in args:
    print(arg)

print_args(1,2,3)
print()
print_args('a', 'b', 'c')
````

````python
def print_kwargs(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} = {value}")

print_kwargs(name="Alice", age=30, city = "New York")
print_kwargs(fruit="Apple", color="Red", price=1.5)
````

````python
# 응용 실습) 앞선 예시를 활용하여 가변인수가 있는 함수를 사용한 코드를 작성 후 공유해봅시다.

menu = {'stake' : 50000,'pasta' : 15000 ,'orange_juice' : 5000,'pizza' : 20000,'wine' : 30000}

def receipt(payment_method, *orders):
    total_amount = 0
    receipt_details = []

    for order in orders:
        menu_name = order[0]
        quantity = order[1]
        price_per_item = menu[menu_name]
        amount = price_per_item * quantity
        total_amount += amount
        receipt_details.append((menu_name, quantity, price_per_item, amount))

    points = total_amount * 0.05

    print("Receipt:")
    for detail in receipt_details:
        print(f"Menu: {detail[0].capitalize()}, Quantity: {detail[1]}, Price per item: {detail[2]}, Total: {detail[3]}")
    print(f"Total Amount: {total_amount}")
    print(f"Points: {points}")
    print(f"Payment Method: {payment_method}")
    print()

# 예제 호출
receipt('Credit Card', ('stake', 2), ('pasta', 3), ('wine', 1))
receipt('Cash', ('pizza', 2), ('orange_juice', 4))

# 출력 예시
# Receipt:
# Menu: Stake, Quantity: 2, Price per item: 50000, Total: 100000
# Menu: Pasta, Quantity: 3, Price per item: 15000, Total: 45000
# Menu: Wine, Quantity: 1, Price per item: 30000, Total: 30000
# Total Amount: 175000
# Points: 8750.0
# Payment Method: Credit Card

# Receipt:
# Menu: Pizza, Quantity: 2, Price per item: 20000, Total: 40000
# Menu: Orange_juice, Quantity: 4, Price per item: 5000, Total: 20000
# Total Amount: 60000
# Points: 3000.0
# Payment Method: Cash
````

## 튜플 인수 언패킹(Unpacking)
튜플 안에 여러 값이 들어 있을 때, 이를 함수의 개별 인수로 분리하여 전달하는 것.

사용방법 : 함수 정의시에는 특별한 문법이 필요없지만, 함수 호출 시 인수 앞에 *를 붙여서 사용합니다.

````python
def display_info(name, age, country):
  print(f"Name:{name}")
  print(f"Age:{age}")
  print(f"Country:{country}")

# 튜플 정의
info=("Alice", 30, "USA")

# 튜플 언패킹을 사용하여 함수 호출
display_info(*info)
````

함수에 딕셔너리를 전달할 때, 딕셔너리 안에 여러 키-값 쌍이 있을 때, 이를 함수의 개별 키워드 인수로 분리하여 전달하는 것.

함수 정의 시에는 특별한 문법이 필요 없지만, 함수 호출 시 인수 앞에 **를 붙여서 사용

## 언패킹(unpacking)

데이터 구조(예 : 리스트, 튜플 등) 안에 묶여 있는 여러 개의 값을 개별적인 변수나 인수로 분리하는 작업

패킹(packing) : 여러 개의 값을 하나의 데이터 구조(예 : 튜플, 리스트 등)로 묶는 것

```
packed = (1,2,3)
```
언패킹(unpacking) : 하나의 데이터 구조 안에 묶여 있는 값을 개별적인 변수나 인수로 분리하는 것
```
a,b,c = packed # 튜플을 언패킹하여 개별 변수에 할당
```
함수 인수에서의 언패킹

함수에 튜플이나 리스트를 전달할 때, 이를 개별 인수로 분리하여 전달하는 것

예를 들어, 하나의 튜플 (1,2,3)을 함수에 전달할 때, * 연산자를 사용하여 이를 1,2,3의 개별 인수로 분리하는 과정이 언패킹.

````python
def print_numbers(*args):
  for arg in args:
    print(arg)

print_numbers(1,2,3)
print_numbers((4,5),6,[7,8])
````

## Built_In Function

### all(iterable)
 - 주어진 iterable(리스트, 튜플, 집합(set), 문자열)의 모든 요소가 참(True)인지 여부를 판단하는 함수
 ```
 print(all[1,2,3]) -> True
 print(all[0,2,3]) -> False
 ```

### any(iterable)
 - iterable의 요소 중 어느 하나라도 참이면 True반환
 - iterable이 비어 있으면 False
 ```
 print(any([1,0,1,0])) -> True
 print(any([0,0,0,0])) -> False
 ```

### bin(x)
- 주어진 정수를 이진 형태의 문자열로 변환하는 함수, 단 앞에 '0b'가 붙는다
```
print(bin(3)) -> '0b11'
print(bin(-10)) -> '-0b1010'
```

### breakpoint(*args, **kws)
 - 프로그램 실행 중에 디버거를 호출하고 해당 지점에서 프로그램 실행을 일시 정지시키는 함수
 ```
 def calculate_sum(n):
  total = 0
  for i in range(n):
    total += i
    if total > 10:
      breakpoint() # 디버깅 지점
    return total
 result = calculate_sum(5)
 print(result)
 -> total이 10을 초과할 때마다 프로그램 실행을 일시 정지, continue명령어를 입력하면 프로그램의 실행이 다시 시작
 ```
 - 하지만 Colab에서는 사용이 어렵고 터미널에서 사용할 수 있다.

### append(), pop()
 - 리스트에 요소를 추가하거나 제거하는 함수
 ```
  fruits = ['apple', 'banana', 'orange']
  fruits.append('grape')
  print(fruits)  ->['apple', 'banana', 'orange', 'grape']

 ```
  popped_fruit = fruits.pop()

  -> 'grape'

### zip()
- 주어진 여러 시퀀스의 요소들을 묶어서 튜플의 리스트로 반환하는 함수(시퀀스의 길이가 다르다면 짧은 시퀀스 길이를 기준)
```

letters = ['a','b','c']
zipped = zip(numbers, letters)
print(list(zipped))
-> [(1,'a'), (2,'b'), (3,'c')]
```

````python
# 1) 변수, 연산문 예제 1
import numpy as np
import matplotlib.pyplot as plt
# 물리 상수
G = 6.67430e-11  # 중력 상수, m^3 kg^-1 s^-2

# 행성의 질량과 초기 위치, 속도
planet_mass = 5.972e24  # kg, 지구의 질량
planet_position = np.array([0.0, 0.0], dtype=float)  # m, 원점에 위치
planet_velocity = np.array([0.0, 0.0], dtype=float)  # m/s, 정지 상태

# 위성의 질량과 초기 위치, 속도
satellite_mass = 1000  # kg
satellite_position = np.array([7.0e6, 0.0], dtype=float)  # m, 지구로부터 7000 km 떨어짐
satellite_velocity = np.array([0.0, 7.12e3], dtype=float)  # m/s, 궤도 속도

# 시뮬레이션 파라미터
time_step = 10  # 초
num_steps = 10000  # 시뮬레이션 스텝 수
def calculate_gravitational_force(m1, m2, pos1, pos2):
    distance_vector = pos2 - pos1
    distance = np.linalg.norm(distance_vector)
    force_magnitude = G * m1 * m2 / distance**2
    force_vector = force_magnitude * distance_vector / distance
    return force_vector

# 궤도 저장을 위한 리스트
planet_positions = []
satellite_positions = []

# 시뮬레이션 루프
for step in range(num_steps):
    force = calculate_gravitational_force(planet_mass, satellite_mass, planet_position, satellite_position)

    # 행성 가속도와 속도 업데이트
    planet_acceleration = force / planet_mass
    planet_velocity += planet_acceleration * time_step
    planet_position += planet_velocity * time_step

    # 위성 가속도와 속도 업데이트
    satellite_acceleration = -force / satellite_mass
    satellite_velocity += satellite_acceleration * time_step
    satellite_position += satellite_velocity * time_step

    # 현재 위치 저장
    planet_positions.append(planet_position.copy())
    satellite_positions.append(satellite_position.copy())
planet_positions = np.array(planet_positions)
satellite_positions = np.array(satellite_positions)

plt.figure(figsize=(8, 8))
plt.plot(planet_positions[:, 0], planet_positions[:, 1], 'bo', label='Planet')
plt.plot(satellite_positions[:, 0], satellite_positions[:, 1], 'ro', label='Satellite')
plt.xlabel('X Position (m)')
plt.ylabel('Y Position (m)')
plt.title('Orbit Simulation')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# 2) 제어문, 함수 예제 2
# 학생 성적 정보 관리
# 학생 정보 -> 딕셔너리에 저장 / for문 사용하여 딕셔너리에 있는 정보 차례대로 출력하는 함수 생성 후 호출
# 딕셔너리 정의
student_grades = {
    '김새싹': {'국어': 85, '영어': 90, '수학': 75},
    '이새싹': {'국어': 92, '영어': 88, '수학': 95},
    '박새싹': {'국어': 78, '영어': 85, '수학': 80},
    '미스터 새싹': {'국어': 90, '영어': 95, '수학': 88},
    '미세스 새싹': {'국어': 85, '영어': 80, '수학': 82}
}

# 성적 정보 함수
def generate_grade_report():
    print("=== 학생 성적 정보 ===\n")
    for student, grades in student_grades.items():
        print(f"이름: {student}")
        print("과목:")
        for subject, grade in grades.items():
            print(f"- {subject}: {grade}")
        average_grade = sum(grades.values()) / len(grades)
        print(f"평균점수: {average_grade:.2f}\n")
        print("----------------------\n")

generate_grade_report()
````

````python
# 3) 종합 예제 등 자유로운 응용 3
# 모래시계모양 별 찍기

def hourglass(n):
    max_width = 2 * n - 1

    # 상단 부분
    for i in range(n):
        print(' ' * i + '*' * (max_width - 2 * i))

    # 하단 부분
    for i in range(1, n):
        print(' ' * (n - i) + '*' * (2 * i))

n = int(input("(5 이상의)숫자를 입력하세요: "))
hourglass(n)
````
