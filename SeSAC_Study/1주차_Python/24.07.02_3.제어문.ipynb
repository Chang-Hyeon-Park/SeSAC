```python
# 가위바위보 게임
# 1) 게임의 규칙 정리
# 2) 필요한 변수 설정 및 초기화(가위 바위 보 : 문자열 또는 숫자로 변경)
# 3) 게임의 시작 및 종료 조건(while, true, break)
# 4) 게임의 점수 조건(변수 설정과 복합연산자)
# 5) 적절한 디버깅

print("가위 바위 보 게임을 시작합니다!\n게임을 종료하고 싶으면 <게임종료> 라고 입력해주세요!\n")

attempts = 0
computer = 0
user = 0
same = 0

while (1):
  rcp = ["가위", "바위", "보"]
  target = random.choice(rcp)

  result = input("가위, 바위, 보, 게임종료 중 하나를 적어주세요! : ")
  if result == "게임종료":
    if computer > user:
      print(f"게임이 종료되었습니다. 컴퓨터 {computer}승 / 당신 {user}승 무승부 {same}번으로 컴퓨터가 이겼습니다!")
    elif computer < user:
      print(f"게임이 종료되었습니다. 컴퓨터 {computer}승 / 당신 {user}승 무승부 {same}번으로 당신이 이겼습니다!")
    elif (computer == 0) and (user == 0) and (same == 0):
      print("게임을 시작하자마자 게임종료를 누르셨군요!")
    elif computer == user:
      print(f"게임이 종료되었습니다. 컴퓨터 {computer}승 / 당신 {user}승 무승부 {same}번으로 비겼습니다!")
    break

  if result not in ["가위","바위","보"]:
    print("가위, 바위, 보, 게임종료 중 하나만을 입력해주세요!\n")
    attempts -= 1

  attempts += 1

  if target == result:
    print(f"비겼습니다! 저는 '{target}'를 냈습니다! 다시 시도해보세요!\n")
    same += 1

  elif target == "가위":
    if result == "보":
      print(f"당신이 졌습니다! 저는 '{target}'를 냈습니다! 다시 해보세요!\n")
      computer += 1
    elif result == "바위":
      print(f"당신이 이겼습니다! 축하합니다!\n{attempts}번의 시도만에 이기셨네요! 저는 '{target}'를 냈습니다!\n")
      user += 1
      attempts = 0

  elif target == "바위":
    if result == "가위":
      print(f"당신이 졌습니다! 저는 '{target}'를 냈습니다! 다시 해보세요!\n")
      computer += 1
    elif result == "보":
      print(f"당신이 이겼습니다! 축하합니다!\n{attempts}번의 시도만에 이기셨네요! 저는 '{target}'를 냈습니다!\n")
      user +=1
      attempts = 0

  elif target == "보":
    if result == "바위":
      print(f"당신이 졌습니다! 저는 '{target}'를 냈습니다! 다시 해보세요!\n")
      computer += 1
    elif result == "가위":
      print(f"당신이 이겼습니다! 축하합니다!\n{attempts}번의 시도만에 이기셨네요! 저는 '{target}'를 냈습니다!\n")
      user +=1
      attempts = 0
```

```python
# 강사님 코드
import random

choices = ["가위", "바위", "보"]
score = {"player": 0, "computer": 0}

while True:
    player = input("가위, 바위, 보 중 하나를 선택하세요 (종료하려면 '끝' 입력): ")
    if player == "끝":
        break

    if player not in choices:
        print("잘못된 입력입니다. 다시 시도하세요.")
        continue

    computer = random.choice(choices)
    print(f"컴퓨터의 선택: {computer}")

    if player == computer:
        print("무승부!")
    elif (player == "가위" and computer == "보") or \
         (player == "바위" and computer == "가위") or \
         (player == "보" and computer == "바위"):
        print("플레이어 승!")
        score["player"] += 1
    else:
        print("컴퓨터 승!")
        score["computer"] += 1

print(f"최종 점수 - 플레이어: {score['player']}, 컴퓨터: {score['computer']}")
```

```python
n = int(input("숫자를 입력하세요: "))

# 별 모양을 위한 패턴 생성
max_width = 2 * n - 1

# 상단 부분
for i in range(n):
    for j in range(n - i - 1):
        print(' ', end='')
    for j in range(2 * i + 1):
        print('*', end='')
    print()

# 중간 부분
for i in range(2 * n - 1):
    if i == n - 1:
        print('*' * (2 * n - 1))
    else:
        print(' ' * (n - 1) + '*')

# 하단 부분
for i in range(n - 1, -1, -1):
    for j in range(n - i - 1):
        print(' ', end='')
    for j in range(2 * i + 1):
        print('*', end='')
    print()
```

```python
def print_star(n):
    max_width = 2 * n - 1

    # 상단 부분
    for i in range(n):
        print(' ' * i + '*' * (max_width - 2 * i) + ' ' * (2 * i) + '*' * (max_width - 2 * i) + ' ' * i)

    # 하단 부분
    for i in range(1, n):
        print(' ' * (n - i) + '*' * (2 * i) + ' ' * (max_width - 2 * i) + '*' * (2 * i) + ' ' * (n - i))

n = int(input("숫자를 입력하세요: "))
print_star(n)
```
