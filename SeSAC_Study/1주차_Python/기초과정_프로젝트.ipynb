그 중 100으로 나누어떨어지는 해는 윤년이 아니다.

400으로 나누어떨어지는 해는 무조건 윤년이다

```python
# 달력 출력기
# import calendar 사용하지 않고 만들어보기

def is_leap_year(y):  # 윤년 계산 함수
  if y % 4 == 0:
    if y % 100 == 0:
      if y % 400 == 0:
        return True
      else:
        return False
    else:
      return True
  else:
    return False


def calendar(year,month):
  if month in [1,3,5,7,8,10,12] :
    day = 31
  elif month in [4,6,9,11]:
    day = 30
  elif month == 2:
    if is_leap_year(year) == True:
      day = 29
    else:
      day = 28

  days_1990_01_01 = 0

  for y in range(1990, year):
    if is_leap_year(y):
      days_1990_01_01 += 366
    else:
      days_1990_01_01 += 365     # 1990년 01월 01일 부터 입력한 년도까지의 일을 더함(윤년 고려)

  for m in range(1,month):
    if m in [1,3,5,7,8,10,12] :
      days_in_month = 31
    elif m in [4,6,9,11]:
      days_in_month = 30
    elif m == 2:
      if is_leap_year(year) == True:
        days_in_month = 29
      else:
        days_in_month = 28

    days_1990_01_01 += days_in_month    # 해당하는 년도에 입력한 월까지의 일을 더함(한 달의 일수 고려)

  # 1990년 1월 1일 월요일을 기준으로 해당하는 달의 시작 날짜의 요일 계산
  start_day = days_1990_01_01 % 7 +1
      # days_1990_01_01 % 7 -> 1990년 1월 1일부터 해당 월과 연도가 시작될때까지 경과된 일 수를
      #                        7로 나눈 나머지를 계산하여 요일 인덱스 제공
      #                        일 0 / 월 1 / ... / 토 6
      # +1 -> 편의상 월요일을 1로 시작하기 위해 +1을 해줌

  # 달력 출력
  print(f"\n\t\t{year}년\t{month:02}월\n") # 1~9일을 01~09로 표시하기 위해 format
  print("일\t월\t화\t수\t목\t금\t토\n")
  print("\t" * start_day, end='') # tab과 start_day를 반복 출력

  for d in range(1, day + 1):
    print(f"{d:02}\t", end="") # 일을 두 자리 수로 format
    if (start_day + d) % 7 == 0: # 한 주가 끝나는걸 계산
      print("\n")
  print()

year = int(input("년도를 입력하세요: "))
month = int(input("월을 입력하세요: "))

calendar(year,month)
```

```python
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

def calendar(year, month):
    # 각 월의 일수 계산
    if month in [1, 3, 5, 7, 8, 10, 12]:
        day = 31
    elif month in [4, 6, 9, 11]:
        day = 30
    elif month == 2:
        if is_leap_year(year):
            day = 29
        else:
            day = 28

    # 요일 계산을 위해 1990년 1월 1일 이후로 경과한 날짜 수를 계산
    days_since_1990_01_01 = 0
    for y in range(1990, year):
        days_since_1990_01_01 += 366 if is_leap_year(y) else 365
    for m in range(1, month):
        days_in_month = 31 if m in [1, 3, 5, 7, 8, 10, 12] else \
                        30 if m in [4, 6, 9, 11] else \
                        29 if is_leap_year(year) else 28
        days_since_1990_01_01 += days_in_month

    # 1990년 1월 1일은 월요일이므로 해당 월의 첫 날 요일 계산
    start_day = (days_since_1990_01_01 % 7 + 1) % 7  # 0: 월요일, 1: 화요일, ..., 6: 일요일

    # 달력 출력
    print(f"{year}년 {month:02}월")  # 월을 두 자리 수로 출력
    print("일\t월\t화\t수\t목\t금\t토")
    print("\t" * start_day, end="")
    for d in range(1, day + 1):
        print(f"{d:02}\t", end="")  # 일을 두 자리 수로 출력
        if (start_day + d) % 7 == 0:
            print()
    print("\n")

# 테스트 예제
year = int(input("년도를 입력하세요: "))
month = int(input("월을 입력하세요: "))
calendar(year, month)
```

```python
import random

# 로또 번호 생성 함수 (스토리 추가)
def generate_lotto_numbers(theme):
    if theme == 'adventure':
        numbers = random.sample(range(1, 46), 6)
        numbers.sort()  # 번호 정렬
        return numbers
    elif theme == 'fantasy':
        # 1부터 50까지의 숫자 중 6개를 선택 (허용되지 않는 숫자도 선택 가능)
        numbers = [random.randint(1, 50) for _ in range(6)]
        return numbers
    elif theme == 'sci-fi':
        # 1부터 60까지의 숫자 중 6개를 선택 (특정 규칙을 따라 선택 가능)
        numbers = []
        while len(numbers) < 6:
            number = random.randint(1, 60)
            if number % 2 == 0:  # 짝수만 선택
                numbers.append(number)
        numbers.sort()  # 번호 정렬
        return numbers
    else:
        raise ValueError("지원되지 않는 테마입니다.")

# 로또 티켓 생성 함수
def generate_lotto_tickets(ticket_count, theme):
    tickets = []
    for _ in range(ticket_count):
        ticket = generate_lotto_numbers(theme)
        tickets.append(ticket)
    return tickets

# 메인 함수
def main():
    print("로또 테마를 선택하세요:")
    print("1. 모험 (Adventure)")
    print("2. 판타지 (Fantasy)")
    print("3. 과학 판타지 (Sci-Fi)")
    choice = input("테마를 선택하세요 (1, 2, 3 중 하나): ")

    if choice == '1':
        theme = 'adventure'
    elif choice == '2':
        theme = 'fantasy'
    elif choice == '3':
        theme = 'sci-fi'
    else:
        print("유효하지 않은 선택입니다.")
        return

    try:
        ticket_count = int(input("생성할 로또 티켓의 수를 입력하세요: "))
        if ticket_count <= 0:
            print("1 이상의 수를 입력하세요.")
            return

        tickets = generate_lotto_tickets(ticket_count, theme)

        print("\n생성된 로또 티켓 번호:")
        for idx, ticket in enumerate(tickets, start=1):
            print(f"티켓 {idx}: {ticket}")
    except ValueError as e:
        print(f"오류 발생: {e}")

if __name__ == "__main__":
    main()
```

```python
import calendar
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

# Define and train your Keras model here
def create_model():
    model = Sequential()
    model.add(Dense(10, input_dim=5, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def train_model(model, X_train, y_train):
    model.fit(X_train, y_train, epochs=10, batch_size=32)

# Save the trained model
def save_model(model, filename):
    model.save(filename)
    print(f"모델이 {filename}에 저장되었습니다.")

def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return False

def print_calendar(year, month):
    # 달력 객체 생성
    cal = calendar.TextCalendar(calendar.SUNDAY)

    # 달력을 문자열로 변환하여 출력
    cal_str = cal.formatmonth(year, month)
    print(cal_str)

# 테스트 예제
year = int(input("년도를 입력하세요: "))
month = int(input("월을 입력하세요: "))
print_calendar(year, month)

# 예제 데이터 생성 (실제 데이터를 사용해야 함)
np.random.seed(0)
X_train = np.random.rand(100, 5)
y_train = np.random.randint(0, 2, size=(100,))

# Save the trained model
model = create_model()
train_model(model, X_train, y_train)
save_model(model, 'my_model.h5')
```

```python
# 3. 달력 출력기(지정된 연도와 월의 달력을 출력)
# ====== 방법1) calendar 사용 =======
import calendar

# 연도와 월을 입력받기
year = int(input("연도를 입력하세요: "))
month = int(input("월을 입력하세요: "))

# 해당 월의 달력 출력
print(calendar.month(year, month))

# ====== 방법2) datetime만 사용 =======
import datetime

def print_calendar(year, month):
    # 매월의 마지막 날짜 list 저장
    days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    # 윤년 확인
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        days_in_month[1] = 29

    # 입력된 월의 1일이 무슨 요일인지 확인 (월요일-일요일 : 0-6)
    # weekday() 메소드 사용
    first_day = datetime.date(year, month, 1).weekday()

    # 해당 월의 총 일수
    num_days = days_in_month[month - 1]

    print(f"{year}년 {month}월의 달력:")
    print("월 화 수 목 금 토 일")

    # 첫 번째 주의 공백 출력
    print("   " * first_day, end="")

    # 날짜 출력
    for day in range(1, num_days + 1):
        print(f"{day:2}", end=" ")
        first_day += 1
        if first_day > 6:
            first_day = 0
            print()

    print()

year = int(input("연도를 입력하세요 (예: 2024): "))
month = int(input("월을 입력하세요 (1-12): "))

print_calendar(year, month)
```
