````python
# 딕셔너리 {key : value}
# 중괄호를 이용해서 딕셔너리 생성, 키(key)와 값(value)의 쌍으로 구성된 자료구조
# 키는 유일(unique)해야 함, 키에 리스트를 사용할 수 없음, 튜플 타입은 사용할 수 있음
# 값은 중복이 가능하며 모든 타입이 가능
# 키 목록에 없는 데이터를 사용하여 참조하면 에러가 발생, 인덱스를 이용한 참조는 지원안함
# items()메소드는 딕셔너리의 모든 키-값 쌍을 튜플 혀앹로 반환합니다.
# keys() 메소드는 딕셔너리의 모든 키를 반환합니다.
# values() 메소드는 딕셔너리의 모든 값을 반환합니다.
````

````python
# 딕셔너리 생성
example_dic = {'name':'Alice', 'age':25, 'city':'Seoul'}

# items() 메서드 사용
print("Items:")
for key, value in example_dic.items():
  print(f"Key : {key}, Value : {value}")

# keys() 메서드 사용
print("\nKeys:")
for key in example_dic.keys():
  print(f"Key : {key}")

# values() 메서드 사용
print("\nValues:")
for value in example_dic.values():
  print(f"Value : {value}")
````

## 실습 1) 학습한 내용 중 다음 내용에 대해 정리하고 다양한 예시코드를 작성하여 공유해 봅시다.

1) 기본 자료형 : int, float, string, bool

2) string의 인덱스 추출

````python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 데이터 생성
x = np.array([[1,2],[2,3],[3,4],[4,5]])
y = np.array([0,0,1,1])

# x변수 : 입력 데이터로, 4개의 샘플과 각 샘플당 2개의 특성을 가진 2차원 배열입니다.
# y변수 : 레이블 데이터로, 각 샘플에 대한 클래스(label)를 나타내는 1차원 배열입니다.
# 여기서는 이진 분류(0 또는 1)로 구성되어 있습니다.


# 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state = 2024)

# x_train 변수 : 학습용 입력 데이터로, 전체 입력 데이터 x의 75%가 할당됩니다.
# x_test 변수 : 테스트용 입력 데이터로, 전체 입력 데이터 x의 25%가 할당됩니다.
# y_train 변수 : 학습용 레이블 데이터로, 전체 레이블 데이터 y의 75%가 할당됩니다.
# y_test 변수 : 테스트용 레이블 데이터로, 전체 레이블 데이터 y의 25%가 할당됩니다.


# 모델 생성 및 학습
model = LogisticRegression()
model.fit(x_train, y_train)

# model 변수 : 로지스틱 회귀 모델 객체 변수에 할당되었습니다.
# model.fit(x_train, y_train) 메서드 : 학습용 데이터(x_train, y_train)를 사용하여 모델을 학습시킵니다.


# 예측 및 평가
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# y_pred 변수 : 테스트용 입력 데이터(x_test)에 대한 모델의 예측 결과입니다.
# accuracy 변수 : 테스트용 레이블 데이터(y_test)와 예측 결과(y_pred)를 비교하여 계산한 모델의 정확도입니다.
# print(f'Accuracy: {accuracy}') : 모델의 정확도를 출력합니다.
````

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy as np

# 데이터 생성
x = np.random.rand(100,10) # 100개의 샘플과 각 샘플당 10개의 특성을 가진 랜덤 숫자로 구성된 2차원 배열
y = np.random.randint(2, size=100) # 0 또는 1의 값을 가지는 100개의 이진 레이블로 구성된 1차원 배열

# 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2024)

# 모델 생성
model = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(1, activation='sigmoid')
])

# model 변수 : 순차적으로 레이어를 쌓아 올리는 Sequential 모델 객체 변수입니다.
# Dense 레이어 : 각 레이어의 노드 수와 활성화 함수, 입력 형태를 정의합니다.
# 첫 번째 Dense 레이어 : 입력 형태가 (10, )이고, 64개의 노드를 가지며, ReLU 활성화 함수를 사용합니다.
# 두 번째 Dense 레이어 : 1개의 노드를 가지며, 시그모이드 활성화 함수를 사용합니다.


# 모델 컴파일
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])

# optimizer : 모델 최적화 방법을 정의하는 변수로, 여기서는 'adam' 옵티마이저를 사용합니다.
# loss : 손실 함수를 정의하는 변수로, 이진 분류 문제를 위해 'binary_crossentropy'를 사용합니다.
# metrics : 모델의 성능을 평가하기 위한 지표를 정의하는 변수로, 여기서는 'accuracy'를 사용합니다.


# 모델 학습
model.fit(x_train, y_train, epochs=10, batch_size=32)

# model.fit() 메서드 : 학습용 데이터(x_train, y_train)를 사용하여 모델을 학습시킵니다.
# epochs 변수 : 모델이 학습 데이터를 반복할 횟수로, 여기서는 10번 반복합니다.
# batch_size 변수 : 한 번의 업데이트에 사용할 샘플 수로, 여기서는 32개씩 사용합니다.


# 모델 평가
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Accuracy: {accuracy}')

# loss 변수 : 테스트 데이터에 대한 손실 값을 저장하는 변수입니다.
# accuracy 변수 : 테스트 데이터에 대한 정확도를 저장하는 변수입니다.
# model.evaluate() 메서드 : 테스트용 데이터(x_test, y_test)를 사용하여 모델의 성능을 평가합니다.
# print(f'Accuracy : {accuracy}') : 모델의 정확도를 출력합니다.
````

## 변수를 생략하거나 _를 사용하는 주요 케이스
_는 특정 값이나 변수를 무시하고자 할 때 자주 사용됩니다.

이는 코드의 가독성을 높이고, 해당 값이 필요하지 않다는 것을 명확히 나타내기 위해 사용됩니다.

## 1. 반복문에서 값 무시하기
for 루프에서 반복 변수의 값을 사용하지 않을 때 _를 사용합니다.

```
# 3번 반복하지만 반복 변수는 사용하지 않음
for _ in range(3):
  print("반복합니다.")
```

## 2. 튜플 언패킹에서 특정 값 무시하기
튜플 언패킹에서 특정 값을 무시하고 싶을 때 _를 사용합니다.

```
# 좌표에서 y값은 무시하고 x,z값만 사용
x,_,z = (1,2,3)
print(f"x:{x},z:{z}")
```

## 3. 여러 반환값 중 일부 무시하기
함수가 여러 값을 반환하지만 일부 값은 필요하지 않을 때 _를 사용합니다.

```
# 함수가 여러 값을 반환하지만 일부 값은 사용하지 않음
def get_coordinaes():
  return (10,20,30)

x,_,z = get_coordinates()
print(f"x:{x},z:{z}"
```

## 4. 인덱스가 필요 없는 enumerate
enumerate를 사용할 때 인덱스가 필요 없는 경우 _를 사용합니다.

```
# 인덱스를 사용하지 않고 리스트의 값만 사용
fruits = ["apple","banana","cherry"]
for _, fruit in enumerate(fruits):
  print(fruit)
```

## 5. 값 무시를 위한 자리 표시자
값을 무시하고 자리를 유지하기 위해 _를 사용합니다.

```
# 두 번째 반환값만 사용
_, value, _ = (1,2,3)
print(value)
```

## 6. 함수 정의에서 사용하지 않는 매개변수
함수 정의에서 사용하지 않는 매개변수에 _를 사용합니다.

```
# 콜백 함수에서 사용하지 않는 매개변수
def callback(_):
  print("콜백이 호출되었습니다.")

callback(None) # 여기서 전달된 값은 사용되지 않음
```

## 7. 리스트 컴프리헨션에서 값 무시하기
리스트 컴프리헨션에서 값 무시하기 위해 _를 사용합니다.

```
# 5개의 None 값을 가진 리스트 생성
none_list = [None for _ in range(5)]
print(none_list)
```

## 변수는 객체에 대한 참조를 저장하는 이름
파이썬에서 변수는 객체 자체를 저장하는 것이 아니라, 객체에 대한 참조(reference)를 저장합니다. 이 참조는 객체의 메모리 주소를 간접적으로 가리키지만, 파이썬 사용자에게는 이러한 세부 사항이 추상화되어 있습니다. 사용자는 변수 이름을 통해 객체에 접근할 수 있습니다.

## 변수(variable)와 객체(object)의 차이
1. 객체 : 메모리에 저장된 데이터입니다. 객체는 값과 관련된 속성(attribute)과 메서드(method)를 가질 수 있습니다. 파이썬의 모든 데이터는 객체입니다. 예를 들어, 숫자, 문자열, 리스트, 함수 등 모두 객체입니다.

2. 변수 : 객체를 참조하는 이름입니다. 변수는 메모리 주소를 통해 객체를 가리킵니다. 즉, 변수는 객체 자체가 아니라 객체에 대한 참조(reference)입니다.
