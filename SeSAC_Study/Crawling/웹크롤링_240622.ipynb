```python
import requests
import bs4
url = 'https://ko.wikipedia.org/wiki/'
query = '파이썬'

req = requests.get(url+query)
soup = bs4.BeautifulSoup(req.text,'html.parser')

# 1. 제목 추출
title = soup.find('caption', class_='infobox-title summary').text

# 2. 이미지 URL 추출
image_url = soup.find('td', class_='infobox-image').find('img')['src']
image_url = 'https:' + image_url  # Making sure the URL is complete

# 3. 설명 문단 추출
description_paragraphs = soup.select('div.mw-parser-output > p')
description = '\n'.join([para.get_text(strip=True) for para in description_paragraphs])

# # 4. 주요 섹션 제목과 내용 추출
sections = soup.find_all('h3')
section_data = {}
for section in sections:
    section_title = section.find('span', class_='mw-headline').text
    section_content = ''
    next_node = section.find_next_sibling()
    while next_node and next_node.name != 'h2':
        if next_node.name == 'p':
            section_content += next_node.get_text(strip=True) + '\n'
        elif next_node.name == 'ul':
            for li in next_node.find_all('li'):
                section_content += '- ' + li.get_text(strip=True) + '\n'
        next_node = next_node.find_next_sibling()
    section_data[section_title] = section_content.strip()

# Display the extracted data
print("Title:", title)
print("Image URL:", image_url)
print("Description:", description)
print("Sections:")
for sec_title, sec_content in section_data.items():
    print(f"\n{sec_title}:\n{sec_content}")
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_email(subject, body, to_email):
    # Gmail SMTP 서버 설정
    smtp_server = 'smtp.gmail.com'
    smtp_port = 587
    username = 'your_email@gmail.com'
    password = 'your_password'

    # 이메일 구성
    msg = MIMEMultipart()
    msg['From'] = username
    msg['To'] = to_email
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'plain'))

    # SMTP 서버에 연결하고 이메일 발송
    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()  # TLS 사용 시작
        server.login(username, password)
        server.send_message(msg)
        server.quit()
        print("Email sent successfully")
    except Exception as e:
        print(f"Failed to send email: {e}")

# 예제 사용법
send_email('Test Subject', 'This is the body of the email.', 'recipient@example.com')
```
