 튜플(tuple)은 소괄호('('와')')를 이용해 생성

 튜플은 읽기 전용, 속도가 발라 수정이 필요 없는 배열 형태의 데이터 타입에 사용

 데이터를 수정할 수 없기 때문에 제공되는 함수가 많지 않음

 튜플의 불변성, 패킹/언패킹 기능, 그리고 여러 값을 하나의 단위로 처리

 데이터가 변경되지 않아야 하는 경우(예 : 딕셔너리의 키)나 성능이 중요한 경우에는 튜플을 사용하고, 데이터를 자주 수정해야 하는 경우에는 리스트를 사용

 데이터의 무결성을 유지, 복합 키가 필요, 함수에서 여러 값을 반환 할 때 특히 유용

# 튜플 vs 리스트
 공통점 : count(), index()
  - 리스트와 튜플 모두에서 사용 가능
  - 이들 메서드는 시퀀스의 내용을 변경하지 않고 정보를 반환하기만 합니다.

  차이점
  - 리스트는 다양한 수정 메서드(append(), extend(), insert(), remove(), pop(), clear(), sort(), reverse())를 가지고 있습니다.
  - 튜플은 불변이기 때문에 내용을 수정하는 메서드가 없습니다.


 - 튜플에서 수정이 필요한 경우, 새로운 튜플을 생성해야 합니다.

```python
def calculate_statistics(numbers):
  mean = sum(numbers) / len(numbers)
  variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
  return mean, variance

# 튜플로 반환된 값을 함수 인수로 사용
numbers = [1,2,3,4,5]
mean, variance = calculate_statistics(numbers)
print(f"Mean: {mean}, Variance: {variance}")

def print_statistics(statistics):
  mean, variance = statistics
  print(f"Mean: {mean}, Variance: {variance}")

# 튜플을 인수로 전달
print_statistics((mean, variance))
```

  - dataset = tf.data.Dataset.from_tensor_slices((features, labels))
   - 이 코드를 실행하면, dataset 객체는 다음과 같은 형태의 요소들을 가지게 됩니다:

# 자료형 2: 딕셔너리 {key:value}
 - 중괄호를 이용해서 딕셔너리 생성, 키와 값의 쌍으로 구성된 자료구조
 - 키는 유일(unique)해야함, 키에 리스트를 사용할 수 없음, 튜플 타입은 사용할 수 있음
 - 값은 중복이 가능하며 모든 타입이 가능
 - 키 목록에 없는 데이터를 사용하여 참조하면 에러가 발생, 인덱스를 이용한 참조는 지원안함

 - 딕셔너리의 키-값 쌍 구조를 활용하여 데이터를 효율적으로 저장하고 접근
 - 딕셔너리는 복잡한 데이터 구조를 표현하거나, 빠른 검색이 필요한 경우, 또는 관계형 데이터를 다룰 때 특히 유용합니다.



 - dictData = {"key" : "value", ...} : 딕셔너리 생성
 - len(dictData) : 딕셔너리의 항목의 수 반환
 - dictData.items() : 딕셔너리의 각 항목들을 (key, value)형식의 튜플들로 반환합니다.
 - dictData.keys() : 딕셔너리의 키(key)들을 반환합니다.
 - dictData.values() : 딕셔너리의 값(value)들을 반환합니다.


 ## List, Dictionary의 공통 매서드
  - clear() : 리스트와 딕셔너리 모두에서 모든 항목을 제거
  - copy() : 리스트와 딕셔너리 모두에서 얕은 복사본을 반환

## 차이가 나는 메서드
 - 리스트와 딕셔너리는 그 구조가 다르기 때문에 대부분의 메서드가 서로 다름
 - 리스트는 순서가 있는 항목의 집합, 딕셔너리는 키-값 쌍의 집합이기 때문

 - 리스트 : append, extend, insert, remove, pop, sort, reverse등
 - 딕셔너리 : items, keys, values, popitem, update등

 - 리스트는 순차적인 데이터 관리에 유리하고,
 - 딕셔너리는 키를 통한 빠른 데이터 접근에 유리


```python
# 새로운 키-값 쌍 추가(insert):
my_dict = {'apple' : 1, 'banana' : 2}
my_dict['cherry'] = 3
print(my_dict)
```

```python
# 여러 개의 새로운 키-값 쌍 추가(insert):
my_dict = {'apple' : 1, 'banana' : 2}
my_dict.update({'cherry' : 3, 'date' : 4})
print(my_dict)
```

```python
# 기존 키의 값 변경(Update):
my_dict = {'apple' : 1, 'banana' : 2}
my_dict['apple'] = 5
print(my_dict)
```

```python
# 기존 데이터의 삭제(del()) : 키값으로 지정
my_dict = {'apple' : 1, 'banana' : 2, 'cherry' : 3}
del my_dict['banana']
print(my_dict)
```

```python
# 기존 데이터의 삭제 (pop()) : 키값으로 지정, 삭제 항목을 리턴값으로 받음
my_dict = {'apple' : 1, 'banana' : 2, 'cherry' : 3}
removed_value = my_dict.pop('banana')
print(my_dict)
print(removed_value)
```

```python
# popitem() : 키값으로 지정, 삭제 항목(키와 밸류값)을 리턴값으로 받음
my_dict = {'apple' : 1, 'banana' : 2, 'cherry' : 3}
last_item = my_dict.popitem()
print(my_dict)
print(last_item)
```

# pop 와 popitem : 요소를 제거

 - pop 메서드:
     - 특정 키를 지정하여 그 키에 해당하는 값을 제거합니다.
     - 제거된 값은 반환됩니다.
     - 만약 지정한 키가 딕셔너리에 없으면 KeyError가 발생합니다
  
  - popitem 매서드:
    - 딕셔너리에서 임의의 키-값 쌍을 제거합니다.
    (파이썬 3.7 이후로는 LIFO 순서로 작동, 즉 마지막에 추가된 항목을 제거함)
    - 제거된 키-값 쌍은 튜플로 반환됩니다.
    - 딕셔너리가 비어있으면 KeyError가 발생합니다.

```python
dictionary = {'a':1,'b':2,'c':3}
value1 = dictionary.pop('b') # value는 2, dictionary는 {'a':1, 'c':3}이 됨
print("Value 1: ", value1)

value2 = dictionary.pop('d','default') # value는 'default', dictionary는 {'a':1, 'c':3} 그대로
print("Value 2: ", value2)
```

```python
dictionary = {'a':1, 'b':2, 'c':3}
item = dictionary.popitem() # item은 ('c',3), dictionary는 {'a':1, 'b':2}이 됨

print("Item: ", item)
```

```python
# 재고 현황 및 가격 현황 시각화

import matplotlib.pyplot as plt
import numpy as np

# 재고 목록 딕셔너리로 선언
inventory = {
    "apple": {
        "price": 1.5,  # 실수형으로 선언
        "stock": 100    # 정수형으로 선언
    },
    "banana": {
        "price": 0.75,
        "stock": 150
    },
    "orange": {
        "price": 1.2,
        "stock": 80
    }
}

# 바나나 재고 50개 증가
inventory['banana']['stock'] += 50

# 오렌지의 가격 출력
print(f"Price of orange: ${inventory['orange']['price']}\n")

# 보조  y축 사용하여 재고와 가격 막대그래프로 시각화
def visualize_inventory():
    products = list(inventory.keys())
    prices = [inventory[product]['price'] for product in products]
    stocks = [inventory[product]['stock'] for product in products]

    fig, ax1 = plt.subplots(figsize=(10, 6))

    bar_width = 0.35
    index = np.arange(len(products))

    rects1 = ax1.bar(index - bar_width/2, stocks, bar_width, label='Stock', color='g')
    ax1.set_xlabel('Products')
    ax1.set_ylabel('Stock', color='g')
    ax1.tick_params(axis='y', labelcolor='g')

    ax2 = ax1.twinx()
    rects2 = ax2.bar(index + bar_width/2, prices, bar_width, label='Price', color='b')
    ax2.set_ylabel('Price ($)', color='b')
    ax2.tick_params(axis='y', labelcolor='b')
    ax2.set_ylim(0, 2.0)  # y축의 값 범위 0~ 2.0으로 설정 (보기 좋게 Scaling)

    ax1.set_xticks(index)
    ax1.set_xticklabels(products)
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    def autolabel(rects, ax):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1, ax1)
    autolabel(rects2, ax2)

    plt.title('Product Inventory')
    plt.tight_layout()
    plt.show()

visualize_inventory()
```

 - 딕셔너리는 키(key)만 셋에 추가 됨
 - 집합 연산자(& : 교집합, | : 합집합, - : 차집합) 사용 가능

## mutable vs immutable

 - Mutable(가변)자료구조
    - 생성 후 내요을 변경할 수 있습니다.
    - 메모리 주소가 변경되지 않습니다.
      - 예 : 리스트(list), 딕셔너리(dict), 집합(set)

 - Immutable(불변) 자료구조
    - 생성 후 내용을 변경할 수 없습니다
    - 내용 변경 시 새로운 객체가 생성됩니다.
      - 예 : 튜플(tuple), 문자열(str), 정수(int), 실수(float)
      

```python
# 리스트 (Mutable)
list_a = [1,2,3]
list_b = list_a
list_b.append(4)
print(list_a)

# 튜플(Immutable)
tuple_a = (1,2,3)
tuple_b = tuple_a
# tuple_b.append(4) # AttributeError: 'tuple' object has no attribute 'append'
tuple_b = tuple_b + (4,)
print(tuple_a)
print(tuple_b)
```

```python
# 얕은 복사와 깊은 복사에서 변경 가능
# 데이터 타입의 복사 방식에 따라 결과가 달라질 수 있습니다.

import copy

# 얕은 복사 (shallow copy)
list_a = [[1,2],[3,4]]
list_b = list_a.copy()
list_b[0][0] = 9
print(list_a)
print(list_b)

# 깊은 복사 (deep copy)
list_c = [[1,2],[3,4]]
list_d = copy.deepcopy(list_c)
list_d[0][0] = 9
print(list_c)
print(list_d)
```

```python
# copy() 는 shallow copy를 생성합니다.
# 즉, 새로운 객체를 만들지만 내부 객체에 대한 참조는 그대로 유지됩니다.

# deepcopy()는 deep copy를 생성합니다.
# 즉, 새로운 객체를 만들고 내부 객체도 모두 새로 생성합니다.
```

```python
# Iterable(반복 가능한 객체)
# 파이썬에서 순회할 수 있는 객체를 의미합니다.
# 요소를 하나씩 반환할 수 있는 객체

# 주요 특징: for 루프에서 사용할 수 있습니다.
# 파이썬의 대표적인 iterable 객체들:
# 리스트(list), 튜플(tuple), 문자열(str),
# 딕셔너리(dict), 집합(set), range 객체
```

```python
# 리스트를 이용한 예시
my_list = [1,2,3,4,5]
for item in my_list:
  print(item, end='')
print()

# 문자열을 이용한 예시
my_string = "Hello"
for char in my_string:
  print(char, end='')
print()

# range 객체를 이용한 예시
for num in range(5):
  print(num, end='')
print()

# 딕셔너리를 이용한 예시
my_dict = {'a':1,'b':2,'c':3}
for key in my_dict:
  print(f"{key}:{my_dict[key]}", end='')
print()

# iterable 객체를 list()로 변환
print(list("Python"))

# 사용자 정의 iterable 사용
class CountDown:
    # 클래스 초기화 메서드. 시작 숫자를 설정합니다.
    def __init__(self, start):
        self.current = start

    # 반복자를 반환하는 메서드. 자신을 반환합니다.
    def __iter__(self):
        return self

    # 다음 값을 반환하는 메서드. 현재 숫자를 반환하고 1 감소시킵니다.
    # 숫자가 0이 되면 StopIteration 예외를 발생시킵니다.
    def __next__(self):
        if self.current > 0:
            num = self.current
            self.current -= 1
            return num
        else:
            raise StopIteration

for num in CountDown(5):
  print(num, end='')
print()
```

```python
# 시작 인덱스 지정
fruits = ['apple','banana','cherry']

for index, fruit in enumerate(fruits, start=1):
  print(f"Index: {index}, Fruit: {fruit}")
```

```python
# List Comprehension와 함께 사용
fruits = ['apple','banana','cheery']
indexed_fruits = [(index, fruit) for index, fruit in enumerate(fruits)]

print(indexed_fruits)
```

```python
# 딕셔너리로 변환
fruits = ['apple','banana','cherry']
fruit_dict = {index: fruit for index, fruit in enumerate(fruits)}
print(fruit_dict)
```

```python
# 문자열에서 문자와 인덱스 출력
text = "hello"

for index, char in enumerate(text):
  print(f"Index: {index}, Character: {char}")
```

```python
# 문자열 형태의 특성을 숫자로 인코딩할 때
# enumerate()를 사용하여 각 문자에 고유한 숫자 인덱스를 부여할 수 있다.
from sklearn.preprocessing import LabelEncoder

labels = ['apple','banana','cherry']
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)

print(list(enumerate(encoder.classes_)))
```

```python
# 모델 평가 시 enumerate()를 사용하여 각 샘플의 인덱스와 예측 결과를 매칭할 수 있습니다.
# 이를 통해 모델의 성능을 샘플 단위로 분석하거나, 오분류된 샘플을 확인할 수 있습니다.
from sklearn.metrics import accuracy_score
y_true = [0,1,0,1,0]
y_pred = [0,0,1,1,0]

for idx, (true, pred) in enumerate(zip(y_true, y_pred)):
  print(f"Sample {idx}: True={true}, Pred={pred}")

print(f"Overall accuracy: {accuracy_score(y_true, y_pred): .2f}")
```
