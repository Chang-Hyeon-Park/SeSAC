## Series(시리즈)
 - 1차원 배열과 유사한 자료 구조로, 데이터와 인덱스(index)가 함께 저장
 - 각 데이터는 인덱스를 통해 접근
 - 인덱스는 기본적으로 0부터 시작, 사용자가 원하는 값으로 설정할 수도
 - 판다스 시리즈는 다양한 데이터 타입을 저장, 데이터 분석에서 많이 사용
 - 시리즈가 1차원 데이터 구조이기 때문에, 데이터프레임의 각 행과 열은 본질적으로 인덱스와 값의 쌍으로 시리즈가 될 수 있다.


s = pd.Series(data, index=index, dtype=dtype, name=name, copy=none)

## 주요 파라미터
 - data : 리스트, 튜플, 딕셔너리, 스칼라값 등
  - 예 : pd.Series([1,2,3,4]) 또는 pd.Series({'a':1,'b':2})
 - index : Series의 라벨 지정
  - 예 : pd.Series([1,2,3], index=['a','b','c'])
 - dtype : 데이터 타입 지정(기본값은 자동 추론)
  - 예 : pd.Series([1,2,3],dtype='float64')
 - name : Series의 이름 지정
  - 예 : pd.Series([1,2,3],name='numbers')
 - copy : 입력 데이터의 복사본 생성 여부(기본값 False)

# 데이터프레임을 병합 : join()
 - 두 개의 데이터프레임을 병합할 때 인덱스를 기준으로 데이터프레임을 결합
 - DataFrame.join(other, on=None, how='left',lsuffix='', rsuffix='', sort=False)
  - ohter : 병합할 다른 데이터 프레임
  - on : 인덱스가 아닌 컬럼을 기준으로 병합할 경우 사용
  - how : 병합 방식('left','right','outer','inner')
  - lsuffix, rsuffix : 병합할 때 컬럼 이름이 중복될 경우 접두사
  - sort : 병합 후 정렬 여부


1. 'left'
 - 왼쪽 데이터프레임(df1)의 모든 행을 유지하면서 오른쪽 데이터프레임(df2)의 일치하는 데이터를 병합. 만약 오른쪽 데이터프레임에 일치하는 데이터가 없다면 NaN으로

2. 'right'
 - 오른쪽 데이터프레임(df2)의 모든 행을 유지하면서 왼쪽 데이터프레임(df1)의 일치하는 데이터를 병합. 만약 왼쪽 데이터프레임에 일치하는 데이터가 없다면 NaN으로

3. 'outer'
 - 두 데이터프레임(df1 과 df2)의 모든 행을 유지하면서 병합합니다. 일치하지 않는 데이터는 NaN으로 채워집니다.

4. 'inner'
 - 두 데이터프레임(df1과 df2)에서 공통으로 존재하는 행만 유지하며 병합합니다. 일치하지 않는 데이터는 제외됩니다.


```python
import pandas as pd

# 예제 데이터프레임 생성
df1 = pd.DataFrame({'A' : ['A0', 'A1', 'A2'],
                    'B' : ['B0', 'B1', 'B2']},
                   index = ['K0','K1','K2'])

df2 = pd.DataFrame({'C' : ['C0', 'C1', 'C2'],
                    'D' : ['D0', 'D1', 'D2']},
                   index = ['K0','K2','K3'])

print(df1)

print(df2)

# join() 사용
result = df1.join(df2, how='outer')
print(result)
```

# 데이터프레임을 병합 : merge() -> 컬럼을 기준으로 병합하는 데 유용

 - pandas.merge(left, right, how='inner',on=None,left_on=None, right_on=None, left_index=False, right_index=False, sort=False)

  - left : 왼쪽 데이터프레임
  - right : 오른쪽 데이터프레임
  - how : 병합 방식('left','right','outer','inner')
  - on : 병합 기준이 되는 컬럼 이름
  - left_on, right_on : 병합 기준이 되는 왼쪽 및 오른쪽 데이터프레임의 컬럼 이름
  - left_index, right_index : 인덱스를 기준으로 병합할지 여부
  - sort : 병합 후 정렬 여부

```python
# 데이터프레임을 병합 : merge(), inner & key(키값이 일치하는 행만!)

import pandas as pd

# 예제 데이터프레임 생성
df1 = pd.DataFrame({'key' : ['K0','K1','K2','K3'],
                    'A' : ['A0', 'A1', 'A2','A3'],
                    'B' : ['B0', 'B1', 'B2','B3']})

df2 = pd.DataFrame({'key' : ['K0','K1','K2','K4'],
                    'C' : ['C0', 'C1', 'C2','C3'],
                    'D' : ['D0', 'D1', 'D2','D3']})

print(df1)

print(df2)

# merge() 사용
result = pd.merge(df1, df2, on='key', how='inner')
print(result)
```

## 원소의 선택
 - [행, 열] 형식의 2차원 좌표로 원소를 지정
 - 행과 열 모두 1개 이상 선택하여 series나 DF 반환 가능

 - 정수 위치 index: DF.iloc[행 번호, 열 번호]

 - df.set_index('이름', inplace=True) : '이름'열을 새로운 인덱스 지정

## 열 데이터의 선택
1. 단일 열 선택
 - 대괄호 표기법 : df['column_name']
 - 점(.) 표기법(공백이 없는 열 이름에만 사용가능): df.column_name

2. 여러 열 선택:
 - 리스트를 사용한 대괄호 표기법 : df[['column1','column2','column3']]
 - 조건을 이용한 열 선택 : df.loc[:, df.columns.str.contains('pattern')]
 - 인덱스 기반 선택(iloc) : df.iloc[:, 0] # 첫 번째 열
  - df.iloc[:, [0,2,4]] # 첫 번째, 세 번째, 다섯 번재 열
  - df.iloc[:, 0:3] # 처음 세 개의 열

 - 라벨 기반 선택(loc)
  - df.loc[:,'column_name']
  - df.loc[:,['column1','column2']]

 - 이름함수를 이용한 선택
  - df.filter(items=['column1','column2'])

 - 불리언 인덱싱
  - mask = [True, False, True, False, ...]
  - DataFrame의 열 수 만큼 df.loc[:, mask]

## 데이터프레임의 결측치 삭제 : dropna()
 - 데이터프레임이나 시리즈에서 결측값(NaN)을 제거하는 데 사용

 DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

 - axis : 결측값을 기준으로 제거할 축을 지정. 0은 행을, 1은 열을 의미. 기본값은 0임
 - how : 'any' 또는 'all'.  'any'는 하나라도 NaN이 있으면 해당 행 또는 열을 제거하고, 'all'은 모든 값이 NaN일 때만 제거. 기본값은 'any'
 - thresh : 이 값보다 많은 비결측값이 있는 행 또는 열만 남긴다.
 - subset : 특정 행이나 열에서만 결측값을 찾아 제거
 - inplace : True 로 설정하면 원본 데이터프레임을 수정

```python
import pandas as pd
import numpy as np

# 예제 데이터프레임 생성
df = pd.DataFrame({
    'A' : [1,2,np.nan, 4],
    'B' : [np.nan, 2,3,4],
    'C' : [1,2,3,np.nan]
})

# 결측값이 있는 행 제거
df_dropped = df.dropna()
print("결측값이 있는 행 제거: \n", df_dropped)
print()

# 모든 값이 NaN인 행 제거
df_dropped_all = df.dropna(how = 'all')
print("모든 값이 NaN인 행 제거: \n", df_dropped_all)
print()

# 특정 열에서만 결측값이 있는 행 제거
df_dropped_subset = df.dropna(subset=['A','B'])
print("특정 열에서만 결측값이 있는 행 제거: \n", df_dropped_subset)
```

# 데이터프레임에 결측값을 채우기 : fillna() 함수
 - 데이터프레임에서 결측값(NaN)을 지정한 값으로 대체하는 데 사용

 DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None)
  - value : 결측값을 대체할 값. 딕셔너리, 시리즈, 스칼라 값 등을 사용할 수 있습니다.
  - method: 'ffill'(앞의 값으로 채우기) 또는 'bfill'(뒤의 값으로 채우기).
  - axis : 결측값을 채울 축. 0 또는 'index'는 행을 따라, 1 또는 'columns'는 열을 따라.
  - inplace : 원본 데이터프레임을 변경할지 여부
  - limit : 결측값을 채울 최대 개수.

```python
import pandas as pd
import numpy as np

# 예제 데이터프레임 생성
df = pd.DataFrame({
    'A' : [1,2,np.nan, 4,5],
    'B' : [np.nan, 2,3,np.nan, 5],
    'C' : [1,np.nan,np.nan,4,5],
    'D' : [np.nan,np.nan,np.nan,np.nan,np.nan]
})

print("원본 데이터프레임:\n", df)
```

```python
# 모든 결측값을 0으로 채우기
df_filled = df.fillna(0)
print("\n모든 결측값을 0으로 채운 결과: \n", df_filled)
```

```python
# 각 열마다 다른 값으로 결측값 채우기
values = {'A' : 10, 'B' : 20, 'C' : 30 , 'D' : 40}
df_filled = df.fillna(value=values)
print("\n각 열마다 다른 값으로 결측값을 채우 결과: \n", df_filled)
```

```python
# apply() 데이터프레임이나 시리즈에 함수를 적용하는 데 사용
import pandas as pd

# 예제 시리즈 생성
s = pd.Series([1,2,3,4,5])

# 각 요소에 2를 곱하는 함수 적용
s_apply = s.apply(lambda x: x*2)

print(s_apply)
```

```python
# apply() : 각 행에 (axis=1)으로 합계를 구하는 함수 적용
import pandas as pd

# 예제 데이터프레임 생성
df = pd.DataFrame({
    'A' : [1,2,3,4],
    'B' : [10,20,30,40]
})

# 각 행에 합계를 구하는 함수 적용
df_apply = df.apply(lambda x: x.sum(), axis=1)

print(df_apply)
```

```python
# apply() : 각 열에 (axis=0)으로 평균을 구하는 함수 적용
import pandas as pd

# 예제 데이터프레임 생성
df = pd.DataFrame({
    'A' : [1,2,3,4],
    'B' : [10,20,30,40]
})

# 각 열에 평균을 구하는 함수 적용
df_apply = df.apply(lambda x: x.mean())

print(df_apply)
```

```python
# apply() : 함수와 열추가를 결합, 열을 추가하는 조건을 설정(열추가 > 요소의 결정)

import pandas as pd

# 예제 데이터프레임 생성
df = pd.DataFrame({
    'A' : [1,2,3,4],
    'B' : [10,20,30,40]
})

# 'A' 열의 값이 2보다 크면 'High', 아니면 'Low'를 반환하는 함수 정의
def categorize(row):
  if row['A'] > 2:
    return 'High'
  else:
    return 'Low'

# 'Category' 열 추가
df['Category'] = df.apply(categorize, axis=1)

print(df)
```

# Bining(구간화)
 - 연속형 데이터를 여러 개의 구간(bin)으로 나누는 과정
 - 데이터의 분포를 분석하거나 시각화할 때 유용
 - pandas에서 주로 cut() 함수와 qcut() 함수를 사용하여 binning을 수행

 - "Bin"이라는 단어는 영어에서 '통','상자','저장소'를 의미
 - "binning"은 연속적인 데이터를 이산적인 구간으로 나누는 과정
 - 연속형 변수를 범주형 변수로 변환

```python
# Bining(구간화)

# cut() 함수
# 데이터를 지정된 구간(bin)으로 나눕니다.
# 비율이 아닌 연속적인 값의 범위에 따라 구간을 나누는 데 사용됩니다.

import pandas as pd

# 예제 데이터
data = [1,7,5,4,6,3,8,9,2,10]

# 데이터프레임 생성
df = pd.DataFrame(data, columns=['Value'])

# 구간 정의
bins = [0,3,6,9,12]

# cut() 함수를 사용하여 데이터 binning
df['Binned'] = pd.cut(df['Value'], bins)

print(df)
```

```python
# qcut() 함수
# 데이터를 동일한 개수의 관측치가 들어가도록 구간(bin)으로 나눕니다.
# 각 구간에 들어가는 데이터의 개수가 거의 동일하도록 나눕니다.

import pandas as pd

# 예제 데이터
data = [1,7,5,4,6,3,8,9,2,10]

# 데이터 프레임 생성
df = pd.DataFrame(data, columns=['Value'])

# qcut() 함수를 사용하여 데이터 binning (4개의 구간으로 나눔)
df['Binned'] = pd.qcut(df['Value'],4)

print(df)
```
